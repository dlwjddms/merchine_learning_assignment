# K-means clustering

## import library

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from matplotlib import cm
```

## load data

```python
fname_data = 'assignment_11_data.csv'

feature = np.genfromtxt(fname_data, delimiter=',')

x = feature[:,0]
y = feature[:,1]

number_data     = np.size(feature, 0)
number_feature  = np.size(feature, 1)

print('number of data : {}'.format(number_data))
print('number of feature : {}'.format(number_feature))
```

## plot the input data

```python
plt.figure(figsize=(8,8))   
plt.title('input data')

plt.scatter(x, y, color='blue')

plt.tight_layout()
plt.show()
```

## compute distance
- feature : $n \times m$, center : $1 \times m$, distance : $n \times 1$
- $n$ : number of data, $m$ : number of features

```python
def compute_distance(feature, center):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    distance = np.array([np.sum(np.power((np.array(center)-np.array(f)),2)) for f in feature])


    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return distance
#compute_distance([[1,2],[3,4],[6,7]],[1,1])
```

## compute centroid
- feature : $n \times m$, label_feature : $n \times 1$, value_label : $1 \times 1$, centroid : $1 \times m$
- $n$ : number of data, $m$ : number of features

```python
def compute_centroid(feature, label_feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    feature = np.array(feature)
    label_feature = np.array(label_feature)
  
    centroid = np.mean(feature[np.where(label_feature==label)], axis=0)
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return centroid
#compute_centroid(np.array([[1,2],[3,4],[5,6]]),np.array([0,1,0]),0)
```

## compute label
- distance : $n \times k$, label_feature : $n \times 1$
- $n$ : number of data, $k$ : number of clusters

```python
def compute_label(distance):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    distance = np.array(distance)
    #print(distance)
    label_feature = np.argmin(distance ,axis=0)
    #print(label_feature)
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return label_feature
#print(np.array([[1,2,3],[3,2,1]]).shape)
#compute_label([[1,2,3],[3,2,1]])
```

## the number of clusters $K = 2$

```python
number_cluster          = 2
number_iteration        = 100     # you can modify this value
loss_iteration_02       = np.zeros(number_iteration)
centroid_iteration_02   = np.zeros((number_iteration, number_cluster, number_feature))
label_feature_02        = np.random.randint(0, number_cluster, size=(number_data))
```

```python
# ++++++++++++++++++++++++++++++++++++++++++++++++++
# complete the blanks
#
for i in range(0, number_iteration):
    for label in range(0, number_cluster):
        if len(np.extract(label_feature_02==label, label_feature_02)):
            centroid_iteration_02[i,label,:] = compute_centroid(feature, label_feature_02, label) 
        else: centroid_iteration_02[i,label,:] =  centroid_iteration_02[i-1,label,:]
   
    loss_iteration_02[i] = np.sum(np.array([np.sum(compute_distance(feature[np.where(label_feature_02==ind)], center)) for ind, center in enumerate(centroid_iteration_02[i])]))/number_data

  
    label_feature_02 = compute_label(np.array([compute_distance(feature, center) for center in centroid_iteration_02[i]]))
    #print(label_feature_02)
    #print( loss_iteration_02[i])
    #if i > 0: 
     #   if np.all(centroid_iteration_02[i-1,:]== centroid_iteration_02[i,:]): break
#
# ++++++++++++++++++++++++++++++++++++++++++++++++++
```
